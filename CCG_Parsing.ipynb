# Import/Config

```python
from lambeq.text2diagram import BobcatParser
parser = BobcatParser(verbose='text')
```

```python
tree = parser.sentence2tree("What is the capital of France?")
```

```python
print(tree.deriv(
    # use_ascii=True,
    vertical=True
))
```
